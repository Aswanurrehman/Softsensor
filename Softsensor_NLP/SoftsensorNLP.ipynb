```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
organ_names=[]
content__=[]
```

```python
url='https://www.pathologyoutlines.com/'
```

```python
r=requests.get(url)
```

```python
print(r.content)
```

```python
soup=BeautifulSoup(r.content,"html.parser")
```

```python
soup.title
```

```python
content=soup.find('div',{"class":"home_content clearfix"})
```

```python
links=content.find_all('a')
```

```python
for i in links:
    #print(i.get('href'))
    organ_names.append(i.text)
    link2=i.get('href')
    pathology=requests.get(link2)
    soup=BeautifulSoup(pathology.text,"html.parser")
    content=soup.find_all('div',{"class":"page_content"})
    content_=[]
    for i in range(len(content)):
        content1=content[i].find_all('div',{"class":"toc_section toc_links"})
    #print(len(content))
        for i in range(len(content1)):
        #print(i)
            content2=content1[i].find_all('span',{"class":"toc_subsection toc_links"})
            for i in range(len(content2)):
                content3=content2[i].find_all('a')
                
                for i in range(len(content3)):
                    #print(content3[i].text)
                    content_.append(content3[i].text)
    content__.append(content_)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(list(zip(organ_names,content__)),columns = ['Organ Name', 'Content'])
```

```python
dictionary=df.to_dict()
```
